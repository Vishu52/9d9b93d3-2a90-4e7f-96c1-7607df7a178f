```python
# Advance Levek Question

# =========================================================
# 21. Sum of digits of numbers in a list (list comprehension)
# =========================================================
print("21. Sum of digits:")
numbers = [123, 45, 908]
digit_sums = [sum(int(d) for d in str(num)) for num in numbers]
print("Sum of digits:", digit_sums)
print("\n")

# =========================================================
# 22. Prime factors using for loop + list comprehension
# =========================================================
print("22. Prime factors:")
num = 60
prime_factors = []
for i in range(2, num + 1):
    if num % i == 0:
        if all(i % j != 0 for j in range(2, i)):
            prime_factors.append(i)
print("Prime factors:", prime_factors)
print("\n")

# =========================================================
# 23. Unique elements from a list (list comprehension)
# =========================================================
print("23. Unique elements:")
lst = [1, 2, 2, 3, 4, 4, 5]
unique = []
[unique.append(x) for x in lst if x not in unique]
print("Unique:", unique)
print("\n")

# =========================================================
# 24. Palindromic numbers up to limit (list comprehension)
# =========================================================
print("24. Palindromic numbers up to 200:")
limit = 200
palindromes = [num for num in range(1, limit + 1) if str(num) == str(num)[::-1]]
print("Palindromes:", palindromes)
print("\n")

# =========================================================
# 25. Flatten a nested list (list comprehension)
# =========================================================
print("25. Flatten nested list:")
nested = [[1, 2], [3, 4], [5, 6]]
flattened = [item for sublist in nested for item in sublist]
print("Flattened:", flattened)
print("\n")

# =========================================================
# 26. Sum of even and odd numbers separately
# =========================================================
print("26. Sum of even & odd separately:")
numbers = [1, 2, 3, 4, 5, 6]
even_sum = sum([n for n in numbers if n % 2 == 0])
odd_sum = sum([n for n in numbers if n % 2 != 0])
print("Even sum:", even_sum)
print("Odd sum:", odd_sum)
print("\n")

# =========================================================
# 27. Squares of odd numbers (1–10)
# =========================================================
print("27. Squares of odd numbers 1–10:")
squares_odd = [n*n for n in range(1, 11) if n % 2 != 0]
print("Squares:", squares_odd)
print("\n")

# =========================================================
# 28. Combine two lists into dictionary
# =========================================================
print("28. List to dictionary:")
keys = ['a', 'b', 'c']
values = [10, 20, 30]
dictionary = {keys[i]: values[i] for i in range(len(keys))}
print("Dictionary:", dictionary)
print("\n")

# =========================================================
# 29. Extract vowels from string (list comprehension)
# =========================================================
print("29. Extract vowels:")
text = "Hello Python World"
vowels = "aeiouAEIOU"
vowel_list = [ch for ch in text if ch in vowels]
print("Vowels:", vowel_list)
print("\n")

# =========================================================
# 30. Remove non-numeric characters from list of strings
# =========================================================
print("30. Remove non-numeric characters:")
strings = ["a1b2", "3x4y", "56z"]
cleaned = [''.join([ch for ch in s if ch.isdigit()]) for s in strings]
print("Numeric only:", cleaned)
print("\n")
```

```python
# challenge Question

# =========================================================
# 31. Sieve of Eratosthenes using list comprehension
# =========================================================
print("31. Prime numbers using Sieve:")
limit = 100
sieve = [True] * (limit + 1)

for i in range(2, int(limit ** 0.5) + 1):
    if sieve[i]:
        sieve[i*i : limit+1 : i] = [False] * len(range(i*i, limit+1, i))

primes = [i for i in range(2, limit + 1) if sieve[i]]
print("Primes:", primes)
print("\n")

# =========================================================
# 32. Pythagorean triplets up to limit
# =========================================================
print("32. Pythagorean Triplets:")
limit = 50
triplets = [(a, b, c) for a in range(1, limit)
                        for b in range(a, limit)
                        for c in range(b, limit)
                        if a*a + b*b == c*c]
print("Triplets:", triplets)
print("\n")

# =========================================================
# 33. All combinations of two lists
# =========================================================
print("33. Combinations of two lists:")
list1 = [1, 2, 3]
list2 = ['a', 'b']
combinations = [(x, y) for x in list1 for y in list2]
print("Combinations:", combinations)
print("\n")

# =========================================================
# 34. Mean, Median, Mode using list comprehension
# =========================================================
print("34. Mean, Median, Mode:")
nums = [1, 2, 2, 3, 4, 4, 4, 5]

mean = sum(nums) / len(nums)

sorted_nums = sorted(nums)
median = sorted_nums[len(nums)//2] if len(nums) % 2 != 0 else \
         (sorted_nums[len(nums)//2 - 1] + sorted_nums[len(nums)//2]) / 2

mode = max(set(nums), key=lambda x: nums.count(x))

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("\n")

# =========================================================
# 35. Pascal's Triangle using list comprehension
# =========================================================
print("35. Pascal's Triangle:")
rows = 6
triangle = []

for i in range(rows):
    if i == 0:
        triangle.append([1])
    else:
        prev = triangle[i - 1]
        row = [1] + [prev[j] + prev[j + 1] for j in range(len(prev)-1)] + [1]
        triangle.append(row)

for row in triangle:
    print(row)
print("\n")

# =========================================================
# 36. Sum of digits of factorial (1 to 5)
# =========================================================
print("36. Sum of digits of factorial:")
import math
results = [sum(int(d) for d in str(math.factorial(n))) for n in range(1, 6)]
print("Digit sums:", results)
print("\n")

# =========================================================
# 37. Longest word in a sentence
# =========================================================
print("37. Longest word:")
sentence = "Python makes data science powerful and fun"
words = sentence.split()
longest = max(words, key=lambda w: len(w))
print("Longest word:", longest)
print("\n")

# =========================================================
# 38. Filter strings with more than 3 vowels
# =========================================================
print("38. Strings having more than 3 vowels:")
words = ["beautiful", "sky", "queue", "python", "education"]
vowels = "aeiouAEIOU"

filtered = [w for w in words if sum(ch in vowels for ch in w) > 3]
print("Filtered:", filtered)
print("\n")

# =========================================================
# 39. Sum of digits of numbers 1 to 1000
# =========================================================
print("39. Sum of digits from 1 to 1000:")
digit_sums = [sum(int(d) for d in str(n)) for n in range(1, 1001)]
print("Total sum:", sum(digit_sums))
print("\n")

# =========================================================
# 40. Prime palindromic numbers
# =========================================================
print("40. Prime Palindromes:")
limit = 500
primes = [n for n in range(2, limit)
          if all(n % i != 0 for i in range(2, int(n**0.5)+1))]

prime_palindromes = [p for p in primes if str(p) == str(p)[::-1]]
print("Prime Palindromes:", prime_palindromes)
print("\n")
```
