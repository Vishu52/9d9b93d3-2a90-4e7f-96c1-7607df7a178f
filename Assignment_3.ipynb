```python
# Reduce Question

# -----------------------------------------
# 1. What is reduce() and basic usage
# -----------------------------------------
"""
reduce() applies a function to elements of a sequence cumulatively.
It is available in the 'functools' module.
"""

from functools import reduce

# Example: Sum of list using reduce
nums = [1, 2, 3, 4]
result = reduce(lambda a, b: a + b, nums)
print("Reduce Sum Example:", result)


# -----------------------------------------
# 2. Product of all elements using reduce()
# -----------------------------------------
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda a, b: a * b, numbers)
print("Product of List:", product)


# -----------------------------------------
# 3. Maximum element using reduce()
# -----------------------------------------
num_list = [10, 45, 7, 89, 32]

max_value = reduce(lambda a, b: a if a > b else b, num_list)
print("Maximum Value:", max_value)


# -----------------------------------------
# 4. Concatenate a list of strings
# -----------------------------------------
words = ["Hello", " ", "World", "!"]
sentence = reduce(lambda a, b: a + b, words)
print("Concatenated String:", sentence)


# -----------------------------------------
# 5. Factorial using reduce()
# -----------------------------------------
n = 6
factorial = reduce(lambda a, b: a * b, range(1, n + 1))
print(f"Factorial of {n}:", factorial)


# -----------------------------------------
# 6. GCD using reduce()
# -----------------------------------------
import math

numbers = [84, 48, 36]

gcd_value = reduce(lambda a, b: math.gcd(a, b), numbers)
print("GCD of List:", gcd_value)


# -----------------------------------------
# 7. Sum of digits of a number
# -----------------------------------------
num = 9876
sum_digits = reduce(lambda a, b: a + b, map(int, str(num)))
print("Sum of Digits:", sum_digits)
```

```python
# Filter Question

# -----------------------------------------------------
# 1. Purpose of filter() with an example
# -----------------------------------------------------
"""
filter() is used to filter elements from an iterable based on a
True/False function. It returns only elements for which the function returns True.
"""

nums = [1, 2, 3, 4, 5]

even_check = lambda x: x % 2 == 0

filtered = list(filter(even_check, nums))
print("Filter Example (Even Numbers):", filtered)


# -----------------------------------------------------
# 2. Select even numbers using filter()
# -----------------------------------------------------
numbers = [10, 15, 20, 25, 30, 35, 40]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even Numbers:", even_numbers)


# -----------------------------------------------------
# 3. Select names starting with a specific letter
# -----------------------------------------------------
names = ["Alice", "Bob", "Anita", "Charlie", "Arjun"]

letter = "A"
filtered_names = list(filter(lambda name: name.startswith(letter), names))
print(f"Names starting with '{letter}':", filtered_names)


# -----------------------------------------------------
# 4. Select prime numbers using filter()
# -----------------------------------------------------
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

numbers = list(range(1, 50))
prime_numbers = list(filter(is_prime, numbers))
print("Prime Numbers:", prime_numbers)


# -----------------------------------------------------
# 5. Remove None values from a list
# -----------------------------------------------------
data = [10, None, 25, None, 50, None, 100]

clean_data = list(filter(lambda x: x is not None, data))
print("Without None Values:", clean_data)


# -----------------------------------------------------
# 6. Select words longer than given length
# -----------------------------------------------------
words = ["python", "ai", "filter", "fun", "development"]

length = 4
long_words = list(filter(lambda w: len(w) > length, words))
print(f"Words longer than {length} characters:", long_words)


# -----------------------------------------------------
# 7. Select numbers greater than a threshold
# -----------------------------------------------------
values = [5, 12, 7, 25, 3, 40, 18]

threshold = 10
greater_than_threshold = list(filter(lambda x: x > threshold, values))
print(f"Values greater than {threshold}:", greater_than_threshold)
```

```python
# Recursion Question

# ----------------------------------------------------------
# 1. Explain recursion vs iteration
# ----------------------------------------------------------
"""
Recursion:
- A function calls itself.
- Works by breaking a problem into smaller subproblems.
- Needs a BASE CASE to stop infinite calls.
- Uses more memory (function call stack).

Iteration:
- Uses loops (for/while).
- Usually faster and uses less memory.
- Does not require a base case.

Recursion is elegant but must be used with caution.
"""


# ----------------------------------------------------------
# 2. Factorial using recursion
# ----------------------------------------------------------
def factorial(n):
    if n == 0 or n == 1:   # Base case
        return 1
    return n * factorial(n - 1)

print("Factorial(6):", factorial(6))


# ----------------------------------------------------------
# 3. Recursive Fibonacci
# ----------------------------------------------------------
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print("Fibonacci(10):", fibonacci(10))


# ----------------------------------------------------------
# 4. Sum of all elements in a list using recursion
# ----------------------------------------------------------
def sum_list(lst):
    if not lst:  # Base case (empty list)
        return 0
    return lst[0] + sum_list(lst[1:])

print("Sum of List:", sum_list([1,2,3,4,5]))


# ----------------------------------------------------------
# 5. Preventing infinite recursion
# ----------------------------------------------------------
"""
Always include a BASE CASE.
Example:
    if n == 0:
        return
Python also limits recursion depth to prevent infinite loops.
"""


# ----------------------------------------------------------
# 6. GCD using recursion (Euclidean Algorithm)
# ----------------------------------------------------------
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print("GCD(48, 18):", gcd(48, 18))


# ----------------------------------------------------------
# 7. Reverse a string using recursion
# ----------------------------------------------------------
def reverse_string(s):
    if s == "":
        return s
    return reverse_string(s[1:]) + s[0]

print("Reversed String:", reverse_string("python"))


# ----------------------------------------------------------
# 8. Power function (x^n) using recursion
# ----------------------------------------------------------
def power(x, n):
    if n == 0:
        return 1
    return x * power(x, n-1)

print("Power(2,5):", power(2, 5))


# ----------------------------------------------------------
# 9. All permutations of a string using recursion
# ----------------------------------------------------------
def permutations(s):
    if len(s) == 1:
        return [s]

    result = []
    for i in range(len(s)):
        for p in permutations(s[:i] + s[i+1:]):
            result.append(s[i] + p)
    return result

print("Permutations('abc'):", permutations("abc"))


# ----------------------------------------------------------
# 10. Check palindrome using recursion
# ----------------------------------------------------------
def is_palindrome(s):
    if len(s) <= 1:
        return True
    return s[0] == s[-1] and is_palindrome(s[1:-1])

print("Palindrome Check:", is_palindrome("madam"))


# ----------------------------------------------------------
# 11. Generate all combinations of elements in a list
# ----------------------------------------------------------
def combinations(lst):
    if not lst:
        return [[]]
    first = lst[0]
    rest = combinations(lst[1:])

    with_first = [[first] + comb for comb in rest]

    return rest + with_first

print("Combinations of [1,2,3]:", combinations([1,2,3]))
```
