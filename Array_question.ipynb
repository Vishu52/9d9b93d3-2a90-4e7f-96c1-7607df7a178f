```python
# =============================================================
# 1. Check if array contains duplicates
# =============================================================

arr = [1, 2, 4, 2, 5, 9]
print("Contains duplicates:", len(arr) != len(set(arr)))


# =============================================================
# 2. Rotate array right by k steps
# =============================================================

arr = [1,2,3,4,5,6,7]
k = 3 % len(arr)
rotated = arr[-k:] + arr[:-k]
print("Rotated array:", rotated)


# =============================================================
# 3. Reverse array in-place
# =============================================================

arr = [2,4,5,7,9,12]
l, r = 0, len(arr)-1
while l < r:
    arr[l], arr[r] = arr[r], arr[l]
    l += 1
    r -= 1
print("Reversed array:", arr)


# =============================================================
# 4. Find maximum element in array
# =============================================================

arr = [10, 5, 20, 8, 15]
max_val = arr[0]
for n in arr:
    if n > max_val:
        max_val = n
print("Maximum element:", max_val)


# =============================================================
# 5. Remove duplicates from sorted array (in-place)
# =============================================================

arr = [1,1,2,2,2,3,3,4,4,4,5,5]
i = 0
for j in range(1, len(arr)):
    if arr[j] != arr[i]:
        i += 1
        arr[i] = arr[j]
print("Array without duplicates:", arr[:i+1])




```
