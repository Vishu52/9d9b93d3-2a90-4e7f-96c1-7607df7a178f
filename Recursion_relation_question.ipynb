```python
# ---------------------------------------------
# Q1: Compute T(2) for T(n) = 3T(n-1) + 12n, T(0)=5
# ---------------------------------------------

def T(n):
    if n == 0:
        return 5
    return 3*T(n-1) + 12*n

T_2 = T(2)
T_2
```

```python
# ---------------------------------------------
# Q2: Solve Recurrences (Python explanation only)
# ---------------------------------------------
solutions = {
    "T(n) = T(n-1) + c": "Theta(n)",
    "T(n) = 2T(n/2) + n": "Theta(n log n)",
    "T(n) = 2T(n/2) + c": "Theta(n)",
    "T(n) = T(n/2) + c": "Theta(log n)"
}

solutions
```
