```python

# String Based Assignment Problem

import re
import string

print("\n========= STRING ASSIGNMENT SOLUTIONS (ALL 50) =========\n")

s = "Hello World from Data Science"
email = "student@example.com"
url = "https://www.google.com"
csv_string = "name,age,city\nJohn,25,New York\nAnna,30,London"


# 1. Reverse string
print("1.", s[::-1])

# 2. Palindrome check
print("2.", "madam" == "madam"[::-1])

# 3. Uppercase
print("3.", s.upper())

# 4. Lowercase
print("4.", s.lower())

# 5. Vowel count
print("5.", sum(1 for c in s.lower() if c in "aeiou"))

# 6. Consonant count
print("6.", sum(1 for c in s.lower() if c.isalpha() and c not in "aeiou"))

# 7. Remove whitespaces
print("7.", s.replace(" ", ""))

# 8. Length without len()
count = 0
for _ in s:
    count += 1
print("8.", count)

# 9. Contains word
print("9.", "World" in s)

# 10. Replace word
print("10.", s.replace("World", "Universe"))

# 11. Count occurrences of a word
print("11.", s.split().count("Data"))

# 12. First occurrence
print("12.", s.find("Data"))

# 13. Last occurrence
print("13.", s.rfind("Data"))

# 14. Split string
print("14.", s.split())

# 15. Join words
print("15.", " ".join(["Python", "is", "awesome"]))

# 16. Space to underscore
print("16.", s.replace(" ", "_"))

# 17. Starts with?
print("17.", s.startswith("Hello"))

# 18. Ends with?
print("18.", s.endswith("Science"))

# 19. Title case
print("19.", s.title())

# 20. Longest word
print("20.", max(s.split(), key=len))

# 21. Shortest word
print("21.", min(s.split(), key=len))

# 22. Reverse words
print("22.", " ".join(s.split()[::-1]))

# 23. Alphanumeric?
print("23.", "Hello123".isalnum())

# 24. Extract digits
print("24.", "".join(c for c in "A1B2C3" if c.isdigit()))

# 25. Extract alphabets
print("25.", "".join(c for c in "A1B2C3" if c.isalpha()))

# 26. Uppercase count
print("26.", sum(1 for c in s if c.isupper()))

# 27. Lowercase count
print("27.", sum(1 for c in s if c.islower()))

# 28. Swap case
print("28.", s.swapcase())

# 29. Remove a word
print("29.", " ".join(w for w in s.split() if w != "World"))

# 30. Valid email?
print("30.", bool(re.match(r"^[\w\.-]+@[\w\.-]+\.\w+$", email)))

# 31. Email username
print("31.", email.split("@")[0])

# 32. Email domain
print("32.", email.split("@")[1])

# 33. Replace multiple spaces
print("33.", "  Hello   World   ".split())

print("33 fixed:", " ".join("  Hello   World   ".split()))

# 34. Valid URL?
print("34.", bool(re.match(r"^https?://[A-Za-z0-9./\-_%]+$", url)))

# 35. Extract protocol
print("35.", url.split("://")[0])

# 36. Character frequency
freq = {}
for c in s:
    freq[c] = freq.get(c, 0) + 1
print("36.", freq)

# 37. Remove punctuation
print("37.", "Hello!!! How??".translate(str.maketrans("", "", string.punctuation)))

# 38. Only digits?
print("38.", "12345".isdigit())

# 39. Only alphabets?
print("39.", "Python".isalpha())

# 40. String to characters
print("40.", list(s))

# 41. Anagrams?
print("41.", sorted("listen") == sorted("silent"))

# 42. Caesar encode
shift = 3
encoded = ""
for c in "abc":
    encoded += chr((ord(c) - ord('a') + shift) % 26 + ord('a'))
print("42.", encoded)

# 43. Caesar decode
decoded = ""
for c in encoded:
    decoded += chr((ord(c) - ord('a') - shift) % 26 + ord('a'))
print("43.", decoded)

# 44. Most frequent word
words = s.split()
print("44.", max(set(words), key=words.count))

# 45. Unique words
print("45.", list(set(words)))

# 46. Syllable count (vowel count)
print("46.", sum(1 for c in s.lower() if c in "aeiou"))

# 47. Special characters present?
print("47.", any(c in string.punctuation for c in s))

# 48. Remove nth word
n = 2
w = s.split()
if 0 <= n < len(w):
    del w[n]
print("48.", " ".join(w))

# 49. Insert word at nth position
words = s.split()
words.insert(2, "Inserted")
print("49.", " ".join(words))

# 50. CSV to list of lists
csv_output = [row.split(",") for row in csv_string.split("\n")]
print("50.", csv_output)

print("\n========= END =========")
```

```python
# List Based Practice Problem :
import random
import math

print("\n=========== LIST ASSIGNMENT SOLUTIONS (ALL 50) ===========\n")

# 1. Create list 1–10
lst = list(range(1, 11))
print("1.", lst)

# 2. Length without len()
length = 0
for _ in lst:
    length += 1
print("2.", length)

# 3. Append element
a = lst.copy()
a.append(11)
print("3.", a)

# 4. Insert at index
a = lst.copy()
a.insert(3, 99)
print("4.", a)

# 5. Remove by value
a = lst.copy()
a.remove(5)
print("5.", a)

# 6. Remove by index
a = lst.copy()
del a[2]
print("6.", a)

# 7. Element exists?
print("7.", 7 in lst)

# 8. Index of first occurrence
print("8.", lst.index(3))

# 9. Count occurrences
print("9.", [1,2,2,3,2].count(2))

# 10. Reverse list
print("10.", lst[::-1])

# 11. Sort ascending
print("11.", sorted(lst))

# 12. Sort descending
print("12.", sorted(lst, reverse=True))

# 13. Even numbers 1–20
print("13.", [x for x in range(1,21) if x % 2 == 0])

# 14. Odd numbers 1–20
print("14.", [x for x in range(1,21) if x % 2 != 0])

# 15. Sum of list
print("15.", sum(lst))

# 16. Max
print("16.", max(lst))

# 17. Min
print("17.", min(lst))

# 18. Squares 1–10
print("18.", [x*x for x in range(1,11)])

# 19. Random numbers
print("19.", [random.randint(1,50) for _ in range(10)])

# 20. Remove duplicates
print("20.", list(set([1,2,3,2,1])))

# 21. Common elements
print("21.", list(set([1,2,3,4]) & set([3,4,5])))

# 22. Difference
print("22.", list(set([1,2,3]) - set([2,3])))

# 23. Merge lists
print("23.", [1,2,3] + [4,5,6])

# 24. Multiply by 2
print("24.", [x*2 for x in lst])

# 25. Filter even
print("25.", [x for x in lst if x % 2 == 0])

# 26. Strings → ints
print("26.", list(map(int, ["1","2","3","4"])))

# 27. Ints → strings
print("27.", list(map(str, [1,2,3])))

# 28. Flatten nested list
nested = [[1,2],[3,4],[5]]
flat = [x for sub in nested for x in sub]
print("28.", flat)

# 29. First 10 Fibonacci numbers
fib = [0,1]
for i in range(8):
    fib.append(fib[-1] + fib[-2])
print("29.", fib)

# 30. Check sorted
print("30.", lst == sorted(lst))

# 31. Rotate left n
n = 2
print("31.", lst[n:] + lst[:n])

# 32. Rotate right n
print("32.", lst[-n:] + lst[:-n])

# 33. Prime numbers up to 50
primes = []
for num in range(2,51):
    if all(num % i != 0 for i in range(2,int(num**0.5)+1)):
        primes.append(num)
print("33.", primes)

# 34. Split into chunks
chunk = 3
lst2 = list(range(1,11))
chunks = [lst2[i:i+chunk] for i in range(0,len(lst2),chunk)]
print("34.", chunks)

# 35. Second largest
a = [1,4,6,2,9,5]
print("35.", sorted(set(a))[-2])

# 36. Replace with squares
print("36.", [x*x for x in lst])

# 37. List to dict
print("37.", {v:i for i,v in enumerate(lst)})

# 38. Shuffle list
temp = lst.copy()
random.shuffle(temp)
print("38.", temp)

# 39. First 10 factorials
facts = [math.factorial(i) for i in range(1,11)]
print("39.", facts)

# 40. Common element between two lists?
print("40.", bool(set([1,2,3]) & set([3,4,5])))

# 41. Remove all elements
print("41.", [])

# 42. Replace negative with 0
print("42.", [0 if x < 0 else x for x in [1,-3,4,-2,5]])

# 43. String to words
print("43.", "Hello world from python".split())

# 44. Words to string
print("44.", " ".join(["Python","is","fun"]))

# 45. First n powers of 2
n = 10
print("45.", [2**i for i in range(n)])

# 46. Longest string
strings = ["hi","hello","python","data"]
print("46.", max(strings, key=len))

# 47. Shortest string
print("47.", min(strings, key=len))

# 48. First n triangular numbers
n = 10
print("48.", [i*(i+1)//2 for i in range(1,n+1)])

# 49. Check subsequence
def is_subsequence(a,b):
    it = iter(b)
    return all(x in it for x in a)
print("49.", is_subsequence([2,3], [1,2,3,4]))

# 50. Swap two elements
lst3 = [10,20,30,40]
i, j = 1, 3
lst3[i], lst3[j] = lst3[j], lst3[i]
print("50.", lst3)

print("\n=========== END ===========\n")
```

```python
# Tuple Based Practice Problem :
import random
import math

print("\n=========== TUPLE ASSIGNMENT SOLUTIONS (ALL 50) ===========\n")

# 1. Tuple 1–5
t = (1,2,3,4,5)
print("1.", t)

# 2. Third element
print("2.", t[2])

# 3. Length without len()
count = 0
for _ in t:
    count += 1
print("3.", count)

# 4. Count occurrences
print("4.", (1,2,2,3,2).count(2))

# 5. First index of element
print("5.", t.index(3))

# 6. Element exists?
print("6.", 4 in t)

# 7. Tuple → list
print("7.", list(t))

# 8. List → tuple
print("8.", tuple([1,2,3]))

# 9. Unpacking
a,b,c,d,e = t
print("9.", a,b,c,d,e)

# 10. Even numbers
print("10.", tuple(x for x in range(1,11) if x % 2 == 0))

# 11. Odd numbers
print("11.", tuple(x for x in range(1,11) if x % 2 != 0))

# 12. Concatenate
print("12.", (1,2,3) + (4,5,6))

# 13. Repeat tuple
print("13.", (1,2) * 3)

# 14. Check empty
print("14.", len(()) == 0)

# 15. Nested tuple
nested = (1, (2,3), 4)
print("15.", nested)

# 16. Access first element of nested
print("16.", nested[1][0])

# 17. Single element tuple
print("17.", (5,))

# 18. Compare tuples
print("18.", (1,2,3) == (1,2,3))

# 19. Delete tuple
x = (1,2,3)
del x
print("19.", "Deleted successfully")

# 20. Slice tuple
print("20.", t[1:4])

# 21. Max value
print("21.", max(t))

# 22. Min value
print("22.", min(t))

# 23. String → tuple of chars
print("23.", tuple("hello"))

# 24. Tuple chars → string
print("24.", "".join(("h","e","l","l","o")))

# 25. Mixed datatype tuple
print("25.", (1, "hello", 3.5, True))

# 26. Check identical tuples
print("26.", (1,2) == (1,2))

# 27. Sort tuple
print("27.", tuple(sorted((4,2,1,3))))

# 28. Ints → strings
print("28.", tuple(str(x) for x in (1,2,3)))

# 29. Strings → ints
print("29.", tuple(int(x) for x in ("1","2","3")))

# 30. Merge tuples
print("30.", (1,2,3) + (4,5))

# 31. Flatten nested tuple
nested = (1,(2,3),(4,5))
flat = tuple(x for group in nested for x in (group if isinstance(group, tuple) else (group,)))
print("31.", flat)

# 32. First 5 primes
primes = []
for num in range(2,30):
    if all(num % i != 0 for i in range(2,int(num**0.5)+1)):
        primes.append(num)
        if len(primes)==5:
            break
print("32.", tuple(primes))

# 33. Tuple palindrome check
print("33.", (1,2,3,2,1) == (1,2,3,2,1)[::-1])

# 34. Squares 1–5
print("34.", tuple(x*x for x in range(1,6)))

# 35. Filter even numbers
print("35.", tuple(x for x in t if x % 2 == 0))

# 36. Multiply elements by 2
print("36.", tuple(x*2 for x in t))

# 37. Random numbers tuple
print("37.", tuple(random.randint(1,50) for _ in range(10)))

# 38. Check if sorted
print("38.", (1,2,3,4) == tuple(sorted((1,2,3,4))))

# 39. Rotate left n
n = 2
print("39.", t[n:] + t[:n])

# 40. Rotate right n
print("40.", t[-n:] + t[:-n])

# 41. First 5 Fibonacci numbers
fib = [0,1]
for _ in range(3):
    fib.append(fib[-1] + fib[-2])
print("41.", tuple(fib))

# 42. User input tuple (demo with fixed input)
print("42.", tuple("a b c".split()))

# 43. Swap two elements in tuple
temp = list(t)
temp[1], temp[3] = temp[3], temp[1]
print("43.", tuple(temp))

# 44. Reverse tuple
print("44.", t[::-1])

# 45. First n powers of 2
n = 5
print("45.", tuple(2**i for i in range(n)))

# 46. Longest string
strings = ("hi","hello","python","a")
print("46.", max(strings, key=len))

# 47. Shortest string
print("47.", min(strings, key=len))

# 48. First n triangular numbers
n = 5
print("48.", tuple(i*(i+1)//2 for i in range(1,n+1)))

# 49. Tuple subsequence check
def is_subsequence(a,b):
    it = iter(b)
    return all(x in it for x in a)
print("49.", is_subsequence((2,3), (1,2,3,4)))

# 50. Alternating 1s and 0s
n = 10
print("50.", tuple(1 if i % 2 == 0 else 0 for i in range(n)))

print("\n=========== END ===========\n")
```

```python
# Set Based Practice Problem :
import random
import math

print("\n=========== SET ASSIGNMENT SOLUTIONS (ALL 50) ===========\n")

# 1. Create set 1–5
s = {1,2,3,4,5}
print("1.", s)

# 2. Add element
a = s.copy()
a.add(6)
print("2.", a)

# 3. Remove element
a = s.copy()
a.remove(3)
print("3.", a)

# 4. Element exists?
print("4.", 4 in s)

# 5. Length without len()
count = 0
for _ in s:
    count += 1
print("5.", count)

# 6. Clear set
a = s.copy()
a.clear()
print("6.", a)

# 7. Even numbers
print("7.", {x for x in range(1,11) if x % 2 == 0})

# 8. Odd numbers
print("8.", {x for x in range(1,11) if x % 2 != 0})

# 9. Union
print("9.", {1,2,3} | {3,4,5})

# 10. Intersection
print("10.", {1,2,3} & {2,3,4})

# 11. Difference
print("11.", {1,2,3,4} - {3,4})

# 12. Subset
print("12.", {1,2}.issubset({1,2,3}))

# 13. Superset
print("13.", {1,2,3}.issuperset({1,2}))

# 14. Set from list
print("14.", set([1,2,3,3,2]))

# 15. Set to list
print("15.", list({1,2,3}))

# 16. Remove random element
a = {1,2,3,4}
removed = a.pop()
print("16.", removed, a)

# 17. Pop element
a = {10,20,30}
print("17.", a.pop())

# 18. No elements in common
print("18.", {1,2}.isdisjoint({3,4}))

# 19. Symmetric difference
print("19.", {1,2,3} ^ {3,4,5})

# 20. Update set
a = {1,2}
a.update({3,4})
print("20.", a)

# 21. First 5 primes
primes = []
for num in range(2,30):
    if all(num % i != 0 for i in range(2,int(num**0.5)+1)):
        primes.append(num)
        if len(primes)==5:
            break
print("21.", set(primes))

# 22. Identical sets
print("22.", {1,2,3} == {3,2,1})

# 23. Frozen set
print("23.", frozenset([1,2,3]))

# 24. Disjoint check
print("24.", {1,2}.isdisjoint({3,4}))

# 25. Squares 1–5
print("25.", {x*x for x in range(1,6)})

# 26. Filter even numbers
print("26.", {x for x in s if x % 2 == 0})

# 27. Multiply all elements by 2
print("27.", {x*2 for x in s})

# 28. Set of random numbers
print("28.", {random.randint(1,20) for _ in range(10)})

# 29. Check empty
print("29.", len(set()) == 0)

# 30. Nested set using frozenset
nested = {1,2,frozenset({3,4})}
print("30.", nested)

# 31. Discard element (no error if missing)
a = {1,2,3}
a.discard(2)
print("31.", a)

# 32. Compare sets
print("32.", {1,2} == {2,1})

# 33. Set from string
print("33.", set("hello"))

# 34. Set strings → ints
print("34.", {int(x) for x in {"1","2","3"}})

# 35. Set ints → strings
print("35.", {str(x) for x in {1,2,3}})

# 36. Set from tuple
print("36.", set((1,2,3,4)))

# 37. Set to tuple
print("37.", tuple({1,2,3}))

# 38. Max in set
print("38.", max({4,1,9,3}))

# 39. Min in set
print("39.", min({4,1,9,3}))

# 40. Set from user input (simulated)
print("40.", set("a b c".split()))

# 41. Intersection empty?
print("41.", len({1,2} & {3,4}) == 0)

# 42. First 5 Fibonacci numbers
fib = [0,1]
for _ in range(3):
    fib.append(fib[-1] + fib[-2])
print("42.", set(fib))

# 43. Remove duplicates using set
print("43.", list(set([1,2,2,3,3,4])))

# 44. Same elements regardless of count
print("44.", set([1,2,3]) == set([3,2,1,1,2]))

# 45. First n powers of 2
n = 6
print("45.", {2**i for i in range(n)})

# 46. Common elements with list
print("46.", set([1,2,3]) & set([3,4,5]))

# 47. First n triangular numbers
n = 6
print("47.", {i*(i+1)//2 for i in range(1,n+1)})

# 48. Check subset
print("48.", {2,3}.issubset({1,2,3,4}))

# 49. Alternating 1s & 0s of length n
n = 10
print("49.", {1 if i % 2 == 0 else 0 for i in range(n)})

# 50. Merge multiple sets
print("50.", set.union({1,2},{2,3},{3,4},{4,5}))

print("\n=========== END ===========\n")
```
